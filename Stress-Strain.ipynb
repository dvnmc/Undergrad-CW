Stress-Strain Diagram (Materials Science and Engineering)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from openpyxl import load_workbook
wb = load_workbook(r"/Users/devinmcdonald/Desktop/LoadDisplacement.xlsx", data_only=True)
ws = wb.get_sheet_by_name('Ti26Nb-Solutionized-2-Failure')
```

```python
displ = ws['A'][5:7965]  
displ_lst = [displ[x].value for x in range(len(displ))]

force = ws['B'][5:7965] 
force_lst = [force[x].value for x in range(len(force))]

estress = ws['C'][5:7965] 
estress_lst = [estress[x].value for x in range(len(estress))]

estrain = ws['D'][5:7965]  
estrain_lst = [estrain[x].value for x in range(len(estrain))]

tstress = ws['E'][5:7965] 
tstress_lst = [tstress[x].value for x in range(len(tstress))]

tstrain = ws['F'][5:7965]  
tstrain_lst = [tstrain[x].value for x in range(len(tstrain))]
```

```python
plt.xlabel('Strain', fontsize = 14)
plt.ylabel('Stress (MPa)', fontsize = 14)
plt.title('Stress-Strain Diagram', fontsize = 18)
plt.ylim(0,800)
plt.xlim(0,0.35)
plt.xscale('linear')

fig = plt.gcf()
fig.set_size_inches(11,8)


plot = plt.scatter(tstrain_lst, tstress_lst, 
                   color = 'red', marker = '.', s = 10, 
                   alpha = 0.5, edgecolors = 'none')

plot = plt.scatter(estrain_lst, estress_lst, 
                   color = 'blue', marker = '.', s = 10, 
                   alpha = 0.5, edgecolors = 'none')


plt.legend(['True (\u03C3+1)', 'Engineering'], fontsize = 15, markerscale = 6)
plt.savefig(r"/Users/devinmcdonald/Desktop/Stress-Strain.png", bbox_inches='tight')
```

```python
print('Ultimate Stress (True): ', np.max(tstress_lst), 'MPa')
print('Maximum Strain (True):  ', np.max(tstrain_lst))

print('Ultimate Stress (Engineering): ', np.max(estress_lst), 'MPa')
print('Maximum Strain (Engineering):  ', np.max(estrain_lst))
```
