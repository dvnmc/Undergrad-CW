Enthalpy vs. Temperature (Al) Diagram (Materials Science and Engineering)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
T0 = 298
T1 = 934 # Melting Temp of Al (Gaskell)
T2 = 1573
```

```python
y0 = 0 # 0 enthalpy at 298K

y1 = ((20.67)*T0)+(12.38*10**(-3))*((T0**2)/2) 
y2 = ((20.67)*T1)+(12.38*10**(-3))*((T1**2)/2)

ys = y2-y1 # final y for solid Al

trans = ys+10700 # final y for H trans line

y3 = 31.76*T1
y4 = 31.76*T2

yl = ys+10700

yf = ys+10700+(y4-y3) # final y for liquid Al
```

```python
Al_s = plt.plot([T0,T1], [y0, ys], label = 'Solid Aluminum', lw = 4, c='skyblue')

Htrans = plt.plot([T1, T1], [ys, ys+10700], label = 'H trans', lw = 4, c='limegreen')

Al_l = plt.plot([T1,T2], [yl, yf], label = 'Liquid Aluminum', lw = 4, c='orange')



plt.plot([T1, T1], [0, ys], lw = 2, c='red', ls='--') # T melt

TmK_dot = plt.scatter(934,0, label = 'T melt: 934 K', marker = 'v', c='red', s=200)

plt.plot([0, 298], [y0, y0], lw = 2, c='purple', ls='--') # Enthalpy (at 25 deg C)
H25_dot = plt.scatter(10, 0, label = 'Enthalpy at 25 deg C: 0 J/mol', marker = '<', c='purple', s=200)

plt.plot([1573, 0], [yf, yf], lw = 2, c='saddlebrown', ls='--') # Enthalpy (at 1300 deg C)
H1300_dot = plt.scatter(10, yf, label = 'Enthalpy at 1300 deg C: 48990 J/mol', marker = '<', c='saddlebrown', s=200)

plt.title('Enthalpy vs. Temperature (Aluminum)', fontsize = 20)
plt.ylabel('Enthalpy (H, J/mol)', fontsize = 15)
plt.xlabel('Temperature (K)', fontsize = 15)

plt.legend(fontsize=14)

for i in range(1):
    plt.gcf().get_axes()[i].set_xbound(0, 1600)
    plt.gcf().get_axes()[i].set_ybound(-500, 50000)
    
fig = plt.gcf()
fig.set_size_inches(12,12)
```
