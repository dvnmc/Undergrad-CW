**Applied Nuclear Physics**

```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

**A, Z, and N for Z = 30 (Zinc) to Z = 40 (Zirconium)**

```python
# Zinc
Zn_A = np.array([64, 66, 67, 68, 70])
Zn_Z = np.array([30, 30, 30, 30, 30])
Zn_N = Zn_A - Zn_Z

# Gallium
Ga_A = np.array([69, 71])
Ga_Z = np.array([31, 31])
Ga_N = Ga_A - Ga_Z

# Germanium
Ge_A = np.array([70, 72, 73, 74, 76])
Ge_Z = np.array([32, 32, 32, 32, 32])
Ge_N = Ge_A - Ge_Z

# Arsenic
As_A = np.array([75])
As_Z = np.array([33])
As_N = As_A - As_Z

# Selenium
Se_A = np.array([74, 76, 77, 78, 80, 82])
Se_Z = np.array([34, 34, 34, 34, 34, 34])
Se_N = Se_A - Se_Z

# Bromine
Br_A = np.array([79, 81])
Br_Z = np.array([35, 35])
Br_N = Br_A - Br_Z

# Krypton
Kr_A = np.array([78, 80, 82, 83, 84, 86])
Kr_Z = np.array([36, 36, 36, 36, 36, 36])
Kr_N = Kr_A - Kr_Z

# Rubidium
Rb_A = np.array([85, 87])
Rb_Z = np.array([37, 37])
Rb_N = Rb_A - Rb_Z

# Strontium
Sr_A = np.array([84, 86, 87, 88])
Sr_Z = np.array([38, 38, 38, 38])
Sr_N = Sr_A - Sr_Z

# Yttrium
Y_A = np.array([89])
Y_Z = np.array([39])
Y_N = Y_A - Y_Z

# Zirconium
Zr_A = np.array([90, 91, 92, 94, 96])
Zr_Z = np.array([40, 40, 40, 40, 40])
Zr_N = Zr_A - Zr_Z
```

**Chart of all stable nuclei from Zinc to Zirconium**

```python
sns.set_style('whitegrid')

norm = mpl.colors.Normalize(vmin=30,vmax=40)
cmap = 'viridis_r'

fig, ax = plt.subplots()
plt.scatter(Zn_N, Zn_Z, s = 150, label = 'Zinc', c = Zn_Z, norm = norm, cmap = cmap)
plt.scatter(Ga_N, Ga_Z, s = 150, label = 'Gallium', c = Ga_Z, norm = norm, cmap = cmap)
plt.scatter(Ge_N, Ge_Z, s = 150, label = 'Germanium', c = Ge_Z, norm = norm, cmap = cmap)
plt.scatter(As_N, As_Z, s = 150, label = 'Arsenic', c = As_Z, norm = norm, cmap = cmap)
plt.scatter(Se_N, Se_Z, s = 150, label = 'Selenium', c = Se_Z, norm = norm, cmap = cmap)
plt.scatter(Br_N, Br_Z, s = 150, label = 'Bromine', c = Br_Z, norm = norm, cmap = cmap)
plt.scatter(Kr_N, Kr_Z, s = 150, label = 'Krypton', c = Kr_Z, norm = norm, cmap = cmap)
plt.scatter(Rb_N, Rb_Z, s = 150, label = 'Rubidium', c = Rb_Z, norm = norm, cmap = cmap)
plt.scatter(Sr_N, Sr_Z, s = 150, label = 'Strontium', c = Sr_Z, norm = norm, cmap = cmap)
plt.scatter(Y_N, Y_Z, s = 150, label = 'Yttrium', c = Y_Z, norm = norm, cmap = cmap)
plt.scatter(Zr_N, Zr_Z, s = 150, label = 'Zirconium', c = Zr_Z, norm = norm, cmap = cmap)

plt.xlabel('Neutron Number N', fontsize = 18)
plt.xlim(30,60)
plt.xticks(np.arange(30, 60, step = 1))
plt.ylabel('Proton Number Z', fontsize = 18)
plt.ylim(28,42)
plt.yticks(np.arange(28, 42, step = 1))

x, y = range(0,100), range(0,100)
plt.plot(x, y, ls = '--', c = 'gray', label = 'N = Z')

plt.legend(fontsize = 16, loc = 'lower right')

fig.set_size_inches(20,10)
```

```python
s_y = np.array([30, 31, 31, 32, 32, 31, 31, 32, 32, 33, 33, 34, 34, 35, 35, 
                36, 36, 35, 35, 36, 36, 37, 37, 38, 38, 39, 39, 40, 40])
s_x = np.array([39, 38, 39, 38, 39, 40, 41, 40, 43, 42, 43, 42, 45, 44, 45, 
                44, 45, 46, 47, 46, 49, 48, 49, 48, 51, 50, 51, 50, 52])

nr_x = np.array([34, 36, 37, 38, 40, 44, 46, 48, 50, 50, 54, 56, 46, 42, 40])
nr_y = np.array([30, 30, 30, 30, 30, 32, 34, 34, 36, 37, 40, 40, 38, 36, 34])

shield_y = np.array([38, 38, 36, 36, 34, 32, 34, 36, 38])
shield_x = np.array([49, 48, 46, 44, 42, 38, 40, 42, 46])
```

**s-process path from (the unstable) 69Zn to 92Zr**

**Identify any stable nuclei in this region that cannot be reached in the s-process, and note which nuclei are shielded from the r-process**

```python
sns.set_style('whitegrid')

norm = mpl.colors.Normalize(vmin=30,vmax=40)
cmap = 'viridis_r'

fig, ax = plt.subplots()
plt.scatter(Zn_N, Zn_Z, s = 150, label = 'Zinc', c = Zn_Z, norm = norm, cmap = cmap)
plt.scatter(Ga_N, Ga_Z, s = 150, label = 'Gallium', c = Ga_Z, norm = norm, cmap = cmap)
plt.scatter(Ge_N, Ge_Z, s = 150, label = 'Germanium', c = Ge_Z, norm = norm, cmap = cmap)
plt.scatter(As_N, As_Z, s = 150, label = 'Arsenic', c = As_Z, norm = norm, cmap = cmap)
plt.scatter(Se_N, Se_Z, s = 150, label = 'Selenium', c = Se_Z, norm = norm, cmap = cmap)
plt.scatter(Br_N, Br_Z, s = 150, label = 'Bromine', c = Br_Z, norm = norm, cmap = cmap)
plt.scatter(Kr_N, Kr_Z, s = 150, label = 'Krypton', c = Kr_Z, norm = norm, cmap = cmap)
plt.scatter(Rb_N, Rb_Z, s = 150, label = 'Rubidium', c = Rb_Z, norm = norm, cmap = cmap)
plt.scatter(Sr_N, Sr_Z, s = 150, label = 'Strontium', c = Sr_Z, norm = norm, cmap = cmap)
plt.scatter(Y_N, Y_Z, s = 150, label = 'Yttrium', c = Y_Z, norm = norm, cmap = cmap)
plt.scatter(Zr_N, Zr_Z, s = 150, label = 'Zirconium', c = Zr_Z, norm = norm, cmap = cmap)

plt.scatter(39, 30, s = 800, marker = 'o', facecolors = 'none', edgecolors='red', label = 'Start of s-process path')
plt.scatter(52, 40, s = 800, marker = 's', facecolors = 'none', edgecolors='red', label = 'End of s-process path')

plt.scatter(nr_x, nr_y, s = 350, marker = 'x', c ='black', lw = 2, label = 'Not reached in s-process')

plt.scatter(shield_x, shield_y, s = 800, marker = 'D', facecolors = 'none', edgecolors = 'blue', 
            label = 'Shielded from r-process')


plt.xlabel('Neutron Number N', fontsize = 18)
plt.xlim(30,60)
plt.xticks(np.arange(30, 60, step = 1))
plt.ylabel('Proton Number Z', fontsize = 18)
plt.ylim(28,42)
plt.yticks(np.arange(28, 42, step = 1))

x, y = range(0,100), range(0,100)
plt.plot(x, y, ls = '--', c = 'gray', label = 'N = Z')

plt.plot(s_x, s_y, c = 'red')

plt.legend(fontsize = 16, loc = 'lower right', markerscale = 0.5)

fig.set_size_inches(20,10)
```

**• Horizontal (N+1) line segments indicate neutron capture**

**• Diagonal up-left (N-1, Z+1) line segments indicate beta decay**

**• Diagonal down-right (N+1, Z-1) line segments indicate electron capture**

**• For a given isotope with z protons and n neutrons:**

**• If a stable neighbor exists with n + x neutrons and p - x protons, the isotope (with z protons and n neutrons) is shielded from the r-process**
