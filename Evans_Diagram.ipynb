```python
M = plt.plot(M_x, M_y, label = 'M reaction', lw = 2)
H = plt.plot(H_x, H_y, label = 'H reaction', lw = 2)


E_corr_AC = plt.plot([-3, -3], [-0.2, -1.0], color = 'green', ls = '--', lw = 2)
i_corr_AC = plt.plot([-10,-3], [-0.2, -0.2], color = 'green', ls = '--', lw = 2)

E_corr = plt.scatter(-10, -0.2, color = 'green', 
                     s = 200, marker = '<', label = 'Ecorr (activation) = -0.2 V vs SHE')

i_corr = plt.scatter(-3, -1, color = 'green', 
                     s = 200, marker = 'v', label = 'icorr (activation) = 10^-3 A/cm^2')


E_corr_AC = plt.plot([-5, -5], [-0.4, -1.0], color = 'red', ls = '--', lw = 2)
i_corr_AC = plt.plot([-10,-5], [-0.4, -0.4], color = 'red', ls = '--', lw = 2)

E_corr = plt.scatter(-10, -0.4, color = 'red', 
                     s = 200, marker = '<', label = 'Ecorr (lim) = -0.4 V vs SHE')

i_corr = plt.scatter(-5, -1, color = 'red', 
                     s = 200, marker = 'v', label = 'icorr (lim) = 10^-5 A/cm^2')

for i in range(1):
    plt.gcf().get_axes()[i].set_xbound(-10.05, 0)
    plt.gcf().get_axes()[i].set_ybound(-1.01, 0.3)
    
plt.title("Evans Diagram", fontsize = 20)
plt.xlabel('log(i) [A/cm^2]', fontsize = 15)
plt.ylabel('Potential, E [V vs SHE]', fontsize = 15)


plt.legend(fontsize=12)

fig = plt.gcf()
fig.set_size_inches(12,12)
#plt.savefig('/Users/devinmcdonald/Desktop/Q9_Evans_diagram.png', bbox_inches="tight")
```
